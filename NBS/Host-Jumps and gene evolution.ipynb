#MODEL DESCRIPTION : Host and pathogen model genes and genomes. 

The basic units of every evolutionary model are genes, in this work we will be focusing on genetic units which allow a pathogen to interact with a host by encoding proteins which are in charge of allowing the pathogen to invade and gain resources from the host. These *effector* genes are assumed to be already succesful in interacting with the host's cell chemistry (target genes) and do take into account the dynamics related to the host's immune system response. To do so, we define a *universal* finite set of host's target genes $\Omega=\{t_{\mu}\}_{\mu=1}^{K_T}$ with $t_{\alpha}\neq t_{\beta}$ if $\alpha \neq \beta$. 

In this model, we will define a host $H_n$ as any non-empty subset of $\Omega$ for which all the elements are different. This induces the first parameter of the model, namely. The length of the host's target set $L_{H_n}$. By identifying the elements of the target set with integer numbers and given that we impossed the uniqueness condition of every element, we can represent the host genomes as sorted lists (imposing some order, ascending for example) of integers, or as points of a hypercube in 
${\mathbb Z}^{L_{H_n}}$ bounded by $\{z_j< z_{j+1}<K_T\}_{j=1}^{L_{H_n}-1}$. In other words, by sorting the arrays we force every permutation of the list to represent the same target set.

Concerning the pathogen genes we proceed in a similar way to the case of hosts target set and represent the set of effector genes of a given pathogen by arrays of units labeled by an integer number taken from an urn of size $K_P$. For this case however we do not need to worry about repeats or order, in fact the occurence of repeat genes is believed to play a role in genome size evolution so we would like to incorporate that posibility to the model as discussed in sections LATER.  Alongside their label, we now describe some other attributes which are necesary to propose a model of genome size evolution. 

First of we need to introduce gene lengths, which is our main object of study. If $\mathcal{E}=\{e_j\}_{j=1}^{K_P}$ denote the set of effectors present in a given pathogen's genome, then,  each one of these genes is acompannied by a real number  $l_i$ which stands for the number of bases which compose the sequence of that  gene. As it is discussed LATER, the length of a gene  is succeptible to  changes as a consequence of  some evolutionary event taking place which may introduce an innovation into the host population. We are interested in the evolution of the genome size, however, it is not the entire genome we are looking at here, it is only the fraction of the genome which codes and regulates the ability of the pathogen to invade and therefore gain resources from the host. Having established this, then the  length $L_g$  of our effector genome $g$ is obtained by adding over every unit $e_i$: 

$$L_G=l_{e_1}+l_{e_2}+\dots+l_{e_{N_P}} \equiv \Sigma _{k=1}^{N_e}l_{e_k} $$

   The next step is to introduce the way we characterise the interaction between a gene and the host target set. For this we simply define a score between every pair which can be formed by taking elements from the two sets of labels $\Omega$ and $\mathcal{E}$, and then compare every gene in the pathogen with every target in the host. In order to do so, lets first assume that not every pair $(e_k,t_\mu)$ may be compatible, in other words not every strategy inherited from a previous evolutionary process  by the pathogen may be effective to invade a newly encountered host. A very simple way to model this situation is now described.
   
   Let $e_i$ and $t_j$ represent an arbitrary pair of genes taken from their respective pathogen and host sets. We say that the effector is succesful against the target if the number $A_{e_i,t_j}$ equals one, which may or may not occur according with some probability $c$. This defines and adjacency list (and a Matrix for the full set of pathogen units)  for the effector $e_i$ given by: 

$$A_{e_i,t_j}=\left\{ \begin{array}{ll}
1 & \mbox{with  probability $c$}\\
0 & \mbox{with probability  $1-c$}.\end{array} \right.
$$

This definition allows us to construct a list of length equal to that of the host array who we can now employ to assign the effector a dynamical quantity which represents the "efficency" of this solution and which may be improved (or worsen) by the occurence of mutations. Before discussing how mutations and other evolutionary processes can change the way the pathogen benefits from the host, we first introduce the genes scores and 'expression'.  To do so, we basically describe how to set it up at the very beginning of the system's evolutionary trajectory and in section (LATER) we give details on how these change as the system evolve. 

Let $s_{e_i,l}\gt0$, $l=1,\dots,N_P$  represent a set of real numbers drawn from a probability distribution p(S). Then, the score list of an effector $e_i$  respect to a pathogen $\{t_j\}_{j=1}^{N_P}$ is defined as $S_{e_i}=[s_{e_i,1} A_{e_i,t_1},\dots,s_{e_i,N_P}A_{e_i,t_{N_P}}]$. As discussed in the next section, the purpouse of the score lists is to capture how well a gene performs the biochemical set of tasks for which it carries the information and which are established in the gene's adjacency list (eq bla). In this model we constrain the study to effector genes and we will relate the scores to the pathogen's ability to gain resources from the host.

##Fitness, Expression.

In the previous section, we introduced the scores and adjacency lists as attributes of our effector gene modules. By introducing  the scores we are basically  weighting the connections contained in the bipartite structure constitued by the genetic unit and the pathogen's set. This is introduced with the idea of allowing effectors to become good evolutionary 'solutions'  not only if their scores are high or but also or if they are 'well' connected. In other words, effectors can improve (or worsen) its efficiency by gaining these connections (mutating into a different effector) or by improving its scores. In order to capture these scenarios, let $n\equiv e_i$ denote an effector unit, then we define the 'expression' of $n$ as $W_n$: 

$$ W_n=\frac{\alpha}{L_{H_P}c}\Sigma_{j=1}^{L_{H_P}}w_{nj}$$ 


Where $w_{nj}=\frac{s_{n,j}^z}{s_{n,j}^z+s_o^z}$, and  $\{s_{n,j}\}_{j=0}^{L_{H_P}}$ is the list of scores between gene $n\equiv e_i$  and the host H. Then, by collecting every expression an adding up over every element within the genome $W_{e_k}$, gives the  *fitness* of the genome $g$, i.e:

$$W_g={\mathcal{N}}_{Eff}^{-1}\left(W_{e_1}+W_{e_2}+....+W_{e_{N_{Eff}}}\right)$$

Where the constant ${\mathcal{N}}_{Eff}$ stands for the number of effector genes present a the genome. Equations (Wn) and (Wg) complete the characterization of the model genes and genomes, and as mentioned before they constitute dynamical quantities whose changes will drive a given population to follow one or another evolutionary path. We now must comment on why the choice of these particular functional forms for Eqs (Wn)  and (Wg)).  The former is the well known Hill's function, which introduces two parameters $z$ and $s_o$, and it constitutes the way in which we translate the scores into weighted contributions which take  values in the unit interval. In this context, the parameters play the role of a threshold value ($s_o$) around which a smooth transition between a poorly expresse pairwise units into a saturated regime whereas the other one ($n$) regulates the *steepnes* of such transition. The prefactor $cL_{H_P}$ is introduced (see the cubes discusion below (write it first!)) to keep the sum of the order of unity, it is immediate to note that the maximum value every term on the right hand size of Eqn (Wn) is one. On the other hand the mean number of $L_{H_P}$ 'successes' of binomially distributed random variables with probability parameter $c$ is precisely $L_{H_P}c$. Hence the contribution of every effector gene to the *fitness* of the genome may be larger than the mean expected value only if the number of interactions (ones in the adjacency list) is larger than the average and with values below the threshold. 

Now, lets briefly examine an example with the objective of fixing ideas.  For instance, lets assume pool sizes equal to $K_T=1000$ and $K_P=100$. Number of genome units equal to $N_e=3$ and $N_P=5$ for the pathogen and host respectively. Then  possible realisation for these  *genomes* are: 

$$g_P=\{e_{1},e_{2}, e_{2} \}=\{20,99,1\}$$ and $$g_H=\{t_{1},t_{2},t_{3},t_{4},t_{5} \}=\{8,23,56,78,904\}$$

By setting probability of connecting two integers to $c=\frac{1}{2}$ we form three score vectors by following the description above, for example $S_{e_1}=S_{20}$:  $$\begin{array}{lll} S_{20}&=&[s_1A_{20,8},s_2A_{20,23},s_3A_{20,56},s_4A_{20,78},s_5A_{20,904}]\\
&=&[s_1,0,s_3,0,s_5]\end{array}$$

Where the $s_i$'s are positive and drawn from a chosen distribution.

##Pathogen-Host Gene interactions and Hyper-cubic lattices landscapes.

Let $B_L$ the set of binary strings (zeros and ones) of length $L$, it is clear that the number of elements in such set is $2^L$. A nice feature of this set is that all the elements can be generated in *lexicographical* order by simply concatenating to the elements of $B_1=\{0,1\}$ all the elements of the set $B_{L-1}$. Or simbolically: 

$$B_L=B_1+B_{L-1}$$

This sort of recursive embedding is very easy to visualize if we consider the sets of vertices formed by employing every member of the binary strings as the coordinate of a point in $\mathbb{Z}^L$, by doing so we can identify the vertices as those of a hypercube of dimension $L$ (a line, a cube, a tesseract, etc) which always contain the cube of one dimension lower as one of its *faces*. 

It is immediate to see that for these regular grids the number of neighbors to a given node is $3L$ and most importantly it is possible for a walker to reach every point in the cube by performing a finite number of steps.

On the other hand, by recalling the way we defined the adjacency lists (Equation Adj) for a given pathogen lenght it is immediate to recognize the list as an element of the hypercube vertices set. Therefore our effector model genomes can be pictured (See figs blah blah blah) as some configuration of occupied nodes in this regular grid, and the effects of the evolutionary process as dynamical processes in this configurations, for instance, if we include the possibilty of adding or removing one link as the result of a mutation, this could be seen a a diffusive process within the network. However, as we will se soon, the model  requires the knowledge of other quantities in order to fully characterize the genes and genomes changes beyond walks in the network. 

###RESHUFFLING OF THE LANDSCAPE BY PERFORMING A JUMP

As an example of this, and perhaps the most important of the evolutionary events considered in this work are *host jumps*, which basically consist in generating a new host and evaluate how well does a given genome adapts to it. In that case the network of phenotypes remains the same, but the comparison is made against a perhaps even completely unrelated set of  targets, this has a consequence a sudden reshuffle of the occupied vertices of the hypercube.

###COMBINATORICS, GENOME FITNESS AND THE CONSTRAINT IMPOSED BY C.

Finally let us examine in some detail the effects of the parameter $c$ has over our *gene expression* definition and how it does translate ibto the hypercube genotype methaphor. As we shall see in the next section, these effects will play an important role in a number of evolutionary steps. Eqns (adj) and (expression) imply that the number of non-zero elements within an adjacency list is on the average $L_{H_P}c$. On the other hand, for the entire set of $2^{L_{H_P}}$ binary sequences we can treat the number of non zero elements in the sequence as a binomal random variable with parameter $1/2$, i.e.:

$$Pr(X=n)=\frac{1}{2^{L_{H_P}}}\left( \begin{array}{c}L_{H_P}\\ n \end{array} \right)$$

This is also the probability distribution for the sum of ones within a sequence, which is by definition an upper bound  of the sum of the $w_{nj}$'s in Eqn (Wk), (i.e. $\sum w_{nj}<\sum A_{n,j}$). This makes clear the role played by the parameter $c$ and the introduction of the prefactor $L_{H_P}c$, in the definition for the gene expressions (Eqn(Expressions)) and its basically to keep the sums of the rder of unity, but also to determine if a gene is well expressed ($W_n>=1.0$) or no. The fact that that the parameter $c$ makes the average  number of ones within an effector adjacency is larger or less than the mean number for the entire set $B_L$ depending if $c<0.5$ or $c>0.5$ implies which region of the set is accesible at the moment of sampling the initial configuration. Regardless if the interaction scores are well expressed or not, the fact that a gene has a number of non-zero terms whcih contribute to its expression larger than the average is entirely stichastic and can be approximated by:
$$\sum A_{n,j}\simeq cL_{H_P}+L_{H_P}^{\frac{1}{2}}(c(1-c))^{1/2}$$

The main goal of the study is to investigate the dynamics of evolutionary changes induced by changes  in the environment and population bottlenecks by means of host jumps and bottlencks, to do so we link the fitness of the genotypes with the reproductive rate of a populations by postulatins the simplest population dynamics model, namely the malthusian model, described by the linear equations:
$$N_g(t+1)=f(W_g)N_g(t)$$
Where the function $f$ is taken to be any positive, increasing and continuous function, without loss of generality we choose $f(W)=e^{W}-1$.

###Evolutionary dynamics.
It is assumed a population which reproces asexually, the *"adaptive walks"* are the result of mutation events and repeats. Every gene can mutate which may result in an improvement or worsening of its expression, in some extreme cases in the loss of function, or nothing (neutral mutations), also as lenght increases there might be some chance of gene copying (repeat events) which may alter the genome's length, to be more concrete, we have the following set of events:

*   Mutational changes: Let $\mu$ stand for the mutational rate per generation and let $\bf{s}_i$ represent the set of non-zero scores if gene $i$ against a given host. Then, a gene within a genome $r$ with population $N_r$ can change its scores vector by an amount $\bf{\delta s}$ with probablity rate:

$$T_{i,1}= \mu \frac{N_r l_i}{L_r} \left[ e^{\left|1-w_i\right|}   -1 \right]$$
			
   Where each element of the vector $\bf{\delta s}$ is taken from a symmetric probability distribution with a small standard deviation $\sigma$. Every time this step is performed new $w_{ij}$'s are evaluated alongside all the derived quantities up to the new genome fitness, if this new fitness is larger than the current one then this genomes are kept, if not the new genome is just rejected (Remember the overcome time obtained somwehere else earlier.) The choice of this particular transition probability weight is relies in the factor containing the exponential term which the closer $w_i$ gets to one, the smaller the probability for this type of event to occur. 

   *   Repeats: An already present gene in the genome can be copied and kept within  the genome if the length of the molecule gets larger, this duplication process which i see as a transient solution to a situation in which the gene is not necessarily well adapted but it is useful nevertheles so it is likely that after surpasing a *threshold* length the probability of it being copied by some genome assembly error.  
   
$$\begin{array}{ccl}
T_{i,2}&=&N_r\mu_{Rep}F(L_g)\\
&=&\frac{N_r \mu_{Rep}}{2}\left[1+\tanh (L_g-\hat{L})\right]\\
 \end{array}$$
The choice of this "Glauber"-type of transition rate is considered only on phenomenological basis and it introduces two parameters to the model, namely: $\mu_{Rep}$ which is the repeat occurence rate, and $\hat{L}$ which a length treshold, genome length with values beyond this threshold are more likely to obtain genes due to copying already existing units.

The occurrence of this type of event introduces a gene which inherits all the atributes of the parent gene and obviously increases the new genome's length.

*  Silencing: This event basically removes any functionality of the gene and set all the scores (and derived quantities) to zero, this event tries to module the loss of fucntionality due to mutations or random deletions and introduces two parameters $\mu^{\gamma}$ and $\beta$, the former representing the mutation rate (RECALL DISCUSS THE POSSIBILITY OF ADDING A FRACTION OF DELETERIOUS EVENTS), whereas the latter in an "Arrhenius" term which make the occurrence probability weight only noticable for very low values of $W_i$.

$$T_{i,3}=\mu^{\dagger}\frac{N_rl_i}{L_g}e^{-\beta W_i}$$
