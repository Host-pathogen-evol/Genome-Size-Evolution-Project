```python
#!/usr/bin/env python
#INITIALIZATION
import sys
sys.path.append('../codes/')
from pygsl import rng as rn
import math as mth
import numpy as np
import modA as mda
import pickle
##############################
#Constants and Parameters
SEED=987654321
rk=rn.rng()
rk.set(SEED)
LH=10 #HOST LENGTH
Kh=500 # Host Urn Size
Kp=100 # Pathoghen Urn Size
C=0.5
sohill=0.5
nhill=5
hlpt=[sohill,nhill]
lnsigma=0.702693602048
lnmean=1177.42231142
LNPRT=[lnsigma,lnmean]
lohill=2.0*lnmean
mu1=0.000001
mu2=0.000001
mu3=0.000001
MU=[mu1, mu2, mu3]
B=10.0
Lwn=LH*C
Prms=[B,Lwn,lnmean]
NBo=10000.0 #LENGTH FOR REPEATS THIS IS USELES I THINK
Neff=5
b=1.0 #Saturation Parameter, check an tune this
Nh=5000000
NJMPS=10
Nmin=100
##############################
```

```python
#INITIAL SETUP
HP=mda.hpmatrix.HPMATRIX(Kp,Kh,C,rk) #HOST-PATHOGEN CHEMISTRY
Fo=0.0
No=1
po={}
Natt=0

while Fo<1.0:
    ho=mda.newhost.NEWHOST(LH,Kh,rk) #Creates a new host
    po=mda.genomeinit.GENOMEINIT(Neff,Kp,rk,LNPRT,HP,ho,hlpt,lohill)
    Wo=mda.getwksum.GETWKSUM(po,C*len(ho))
    Fo=mth.exp(Wo)-1.0
    Natt+=1

print Natt
print ho
print Fo
mda.prtgenome.PRTGENOME(po)

Hn={}
Nr={}
Gr={}
Fr={}
Tr={}
Lr={}
Nrloc={}
################
sna="jmp"+str(0)+"gen"+str(0)
Nr[sna]=1.0
Nrloc[sna]=[]
Nrloc[sna].append(Nr[sna])
Fr[sna]=Fo
Gr[sna]=po
Lo=mda.getlength.GETLENGTH(po)
Tr[sna]=mda.trates.TRATES(MU,po,Nr[sna],Prms,Lo)
Lr[sna]=Lo
Hn[0]=ho
cnt=0
```

```python
###LOOOOOOOOP
for jmp in range(NJMPS):
    print jmp
    ###########
    if jmp>0:
    #THE ACTUAL JUMP
        cnt=0
        Hn[jmp]=mda.newhost.NEWHOST(LH,Kh,rk) #
        Nrx={}
        Grx={}
        Frx={}                   
        Nrloc={}
        
        for i in Nr.keys():
            if (Nr[i]>Nmin):
                sna="jmp"+str(jmp)+"gen"+str(cnt)
                Nrx[sna]=Nmin
                Grx[sna]=mda.jumpchange.JUMPCHANGE(Gr[i],Hn[jmp],hlpt,lohill)
                Wk=mda.getwksum.GETWKSUM(Grx[sna],C*len(Hn[jmp]))
                Fk=mth.exp(Wk)-1.0
                Frx[sna]=Fk
                Nrloc[sna]=[]
                Nrloc[sna].append(Nrx[sna])
                cnt+=1
        Nr={}
        Nr=Nrx
        Gr={}
        Gr=Grx
        Fr={}
        Fr=Frx
        Tr={}
        Lr={}
        if(len(Nr.keys)==0):
            print("game over")
            break
        else:
            for i in Nr.keys():
                Lo=mda.getlength.GETLENGTH(Gr[i])
                Tr[i]=mda.trates.TRATES(MU,Gr[i],Nr[i],Prms,Lo)
                Lr[i]=Lo
      
    ###########
    NTOT=[]
    RHIST=[]
    Fmax=max(Fr.values())    
    for tngen in range(5000):
        x1=mda.totpop.TOTPOP(Nr)
        NTOT.append(x1)
        x2=mda.gspsum.GSPSUM(Tr)
        x3=mda.gsptime.GSPTIME(x2,rk) 
        Nrx={}
        Nrx=Nr
        Nr=mda.popupdate.POPUPDATE(Fr,Nrx,Nh,b)
    
        if x3<x1:
            nureact=mda.gspreactselect.GSPREACTSELECT(x2,rk,Tr)
            #print nureact
            ng1=nureact[0]
            ng2=nureact[1]
            ng3=nureact[2]
            #print ng1, ng2, ng3
            newgk=mda.genetrans.GENETRANS(ng2,int(ng3),Gr[ng1],Hn[jmp],hlpt,lohill,rk)
            Wk=mda.getwksum.GETWKSUM(newgk,C*len(Hn[jmp]))
            Fk=mth.exp(Wk)-1.0
            if Fk > Fmax:
                cnt+=1
                sna="jmp"+str(jmp)+"gen"+str(cnt)
                Nr[sna]=1.0
                Fr[sna]=Fk
                Gr[sna]=newgk
                Lr[sna]=mda.getlength.GETLENGTH(newgk)
                Nr[ng1]=Nr[ng1]-1.0
                Nrloc[sna]=[]
                Nrloc[sna].append(Nr[sna])
                Fmax=Fk
                RHIST.append(ng3)
        #Lb=mda.getlength.GETLENGTH(newgb)
        #raw_input()
        for j in Nr.keys():
            Nrloc[j].append(Nr[j])
            Tr[j]=mda.trates.TRATES(MU,Gr[j],Nr[j],Prms,Lr[j])
    
    #ADD JUMPS HERE
    #FIX THIS SOON
    ######################
    print len(Nr)
    SrtFr=[]
    Yr={}
    Yr=mda.sortptnf.SORTPTNF(Nrloc,jmp)
     
    for jn in range(len(Fr.keys())):
        qn="jmp"+str(jmp)+"gen"+str(jn)
        SrtFr.append(Fr[qn])
    ######################
    sreacts="datakerneltest/Rhistory"+str(jmp)+".p"
    spops="datakerneltest/PopsJMP"+str(jmp)+".p" 
    spops2="datakerneltest/NtJMP"+str(jmp)+".p" 
    slens="datakerneltest/LenJMP"+str(jmp)+".p"
    sfns="datakerneltest/RatesJMP"+str(jmp)+".p"
    sgens="datakerneltest/GenesJMP"+str(jmp)+".p"
    pickle.dump(Nr,open(spops,"wb"))
    pickle.dump(Lr,open(slens,"wb"))
    pickle.dump(SrtFr,open(sfns,"wb"))
    pickle.dump(Gr,open(sgens,"wb"))
    ##pickle.dump(Tr,open(sreacts,"wb"))
    pickle.dump(Yr,open(spops2,"wb"))
    pickle.dump(RHIST,open(sreacts,"wb"))
    ######
```
