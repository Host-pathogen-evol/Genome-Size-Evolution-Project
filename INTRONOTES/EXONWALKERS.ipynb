```python
from pygsl import rng as rn
import copy
N=1000.0 #Number of genes
No=N #Initial number of exons
Nf=3.0*No #Final number of exons
SEED=987654321
rk=rn.rng()
rk.set(SEED)

K=Nf-No
#K=10
print K
k={}
Nk={}

for i in range(int(K)):
    k[i+1]=0.0

k[1]=No #initial condition
#print k
#raw_input()

for taun in range(1,int(K+1)):
    if ((taun%100)==0):
        print taun
    k2={}
    k2=copy.deepcopy(k)
    Nk[taun]=k2
    
    #print Nk
    #raw_input
    #raw_input()
    sn=0.0
    for i in k.keys():
        sn+=i*k[i]
        
    r=rk.uniform_pos()*sn
    
    sn=0.0
    for i in k.keys():
        sn+=i*k[i]
        if(sn>=r):
            n=i
            break
            
    k[n]=k[n]-1.0
    k[n+1]=k[n+1]+1.0
            
        
    #print r
    #raw_input()
import pickle
name='Exonwalkers.p'
pickle.dump(Nk,open(name,"wb"))
print("DONE!")
del Nk
```
